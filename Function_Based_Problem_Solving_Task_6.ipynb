# Problem Statement 1:Smart Parking Lot Management System


```python
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

def smart_parking_system(capacity, logs):
    current = 0
    peak = 0

    for action in logs:
        if action == "IN":
            current += 1
        elif action == "OUT":
            current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: FULL - Alert!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)

smart_parking_system(capacity, vehicle_logs)

```

# Problem Statement 2:Online Food Delivery Time Estimator

```python
distance = 8
traffic = "High"
weather = "Rainy"

def estimate_delivery_time(distance, traffic, weather):

    time = distance * 5 #base time

    # Traffic delay
    if traffic == "Medium":
        time += 10
    elif traffic == "High":
        time += 15

    # Weather delay
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")



estimate_delivery_time(distance, traffic, weather)
```

#  Problem Statement 3:Movie Theatre Seat Occupancy

```python
total_seats = 200
booked_seats = [1] * 150   # 150 bookings

def seat_occupancy(total_seats, booked_list):

    booked = sum(booked_list)

    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    if occupancy >= 80:
        print("Suggestion: Consider opening additional shows")


seat_occupancy(total_seats, booked_seats)
```

#  Problem Statement 4:Cloud Server Load Classification System

```python
cpu_data = [45, 60, 70, 85, 90]

def classify_server_load(cpu_readings):
    total = 0

    for cpu in cpu_readings:
        total += cpu

    avg_cpu = total / len(cpu_readings)

    print("Average CPU Load:", int(avg_cpu), "%")

    if avg_cpu < 50:
        print("Server Status: Normal")
    elif avg_cpu < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


classify_server_load(cpu_data)
```

#  Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

def classroom_monitor(resource_usage):
    overused = []

    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

classroom_monitor(usage_data)
```

# Problem Statement 6: Online Event Registration Capacity Controller

```python
capacity = 100
registrations = 105

def event_controller(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # Loop through each registration
    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    if waitlisted > 0:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

event_controller(capacity, registrations)
```
