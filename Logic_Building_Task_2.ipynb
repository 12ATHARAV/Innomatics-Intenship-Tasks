# Problem 1: Unique Words in a Sentence

```python
input_sentence = input("Enter a sentence:")

def unique_words(input_sentence):
    words = input_sentence.split()
    unique_words = set(words)

    print("Unique words count: ", len(unique_words))
    print("Unique words: ", unique_words)

unique_words(input_sentence)
```

# Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

def highest_salary(employees):
    for key, value in employees.items():
        if value == max(employees.values()):
            print(f"{key} - {value}")

highest_salary(employees)
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

def min_max_values(lst):
    max_value = max(lst)
    min_value = min(lst)
    print("Maximum value:",max_value)
    print("Minimum value:",min_value)

min_max_values(numbers)
```

# Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

def price_threshold(lst):
    count = 0
    for price in lst:
        if price > 1000:
            count += 1
    print("Products above 1000:", count)

price_threshold(prices)
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

def attendence_percentage(lst):
    present = lst.count("P")
    total = len(lst)
    percentage = (present / total) * 100
    print("Attendance Percentage:", percentage)

attendence_percentage(attendance)
```

# Problem 6: Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

def duplicate_phone_numbers(lst):
    unique_num = set(lst)
    print("Unique phone numbers:", unique_num)

duplicate_phone_numbers(phone_numbers)
```

# Problem 7: Count Character Frequency

```python
input_string = input("Enter a string: ")

def count_character_freq(string):
    freq = {}
    for char in string:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    return freq

count_character_freq(input_string)
```

# Problem 9: Check if a Key Exists in a Dictionary

```python
dictionary = {"Name":"Atharav",
              "Department":"LLM",
              "Role":"CEO",
              "Salary": "$23000000"}

def key_exists(dictionary):
    key = input("Enter a key:")

    if key in dictionary:
        print(f"{key} exists")
    else:
        print(f"{key} does not exists")

key_exists(dictionary)
```

# Problem 10: Calculate Average Marks

```python
marks = [45,49,47,41,48]

def avg_marks(lst):
    total = sum(lst)
    avg = total / len(lst)
    print("Average Marks:", avg)

avg_marks(marks)
```
