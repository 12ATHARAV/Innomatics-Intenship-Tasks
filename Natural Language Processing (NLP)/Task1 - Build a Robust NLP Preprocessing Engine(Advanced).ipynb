```python
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))   #unique stopwords
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # handle empty string
    if text.strip() == "":
        return ""
    
    # lowercasing
    text = text.lower()
    
    # remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # remove numbers
    text = re.sub(r'\d+', '', text)     # \d+ matches all digits (0–9)
    
    # remove punctuation
    text = text.translate(str.maketrans('','',string.punctuation))
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()    # \s+ replaces multiple spaces with a single space
    
    # remove special characters / emojis
    text = re.sub(r'[^\w\s]', '', text)
    
    # handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # handle only emojis/numbers 
    if text == "":
        return ""
    
    #remove very short tokens <=2 , keep meaningful words like 'no', 'not'
    words = text.split()
    filtered_words = [word for word in words if len(word) > 2 or word in ['no', 'not']]
    
    return ' '.join(filtered_words)
    
```

```python
text = "WOW!!! This is soooo goooood!!! Visit http://example.com now or email me@test.com. I have 2, 24 dogs."
print(preprocess_text(text))
```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for i, text in enumerate(sentences, 1):
    cleaned_sentence = preprocess_text(text)
    cleaned_tokens = cleaned_sentence.split()
    
    print(f"{i}. Original Text: {text}")
    print(f"   Cleaned tokens: {cleaned_tokens}")
    print(f"   Cleaned sentence: {cleaned_sentence}")
    print("-" * 60)
```

## Task 4: Token Analytics

```python
def token_analytics(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens
    
    return total_tokens, unique_tokens, avg_length
```

```python
for i, text in enumerate(sentences, 1):
    cleaned_sentence = preprocess_text(text)
    tokens = cleaned_sentence.split()
    
    total, unique, avg_len = token_analytics(tokens)
    
    print(f"{i}. {text}")
    print(f"   Tokens: {tokens}")
    print(f"   Total tokens: {total}")
    print(f"   Unique tokens: {unique}")
    print(f"   Avg. token length: {avg_len}")
    print('-' * 60)
```

## Task 5: Frequency Analysis

<!-- Task 4: Token Analytics -->

```python
from collections import Counter

all_tokens = []

for text in sentences:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    all_tokens.extend(tokens)
    
# count frequencies
word_freq = Counter(all_tokens)

# top 10 most frequent words
top_10 = word_freq.most_common(10)

# top 5 least frequent words
least_5 = sorted(word_freq.items(), key=lambda x: x[1])[:5]

print("Top 10 most frequent words")
for word, count in top_10:
    print(f"{word}: {count}")
    
print('-' * 60)
    
print("Top 5 least frequent words")
for word, count in least_5:
    print(f"{word}: {count}")
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        # preprocess each sentence
        cleaned = preprocess_text(text)
        
        # store cleaned sentence
        clean_sentences.append(cleaned)
        
        # tokenize and store tokens
        tokens = cleaned.split()
        all_tokens.extend(tokens)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
from pprint import pprint  #pretty print
result = full_pipeline(sentences)

pprint(result)
```
