# Problem 1: Smart Light Controller

```python
class smart_light_controller:
    def __init__(self, light_name):
        self.light_name = light_name
        self.action = "OFF" #default

    def turn_on(self):
        self.action = "ON"

    def turn_off(self):
        self.action = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.action}")

l1 = smart_light_controller("Bedroom Light")
l1.turn_on()
l1.display_status()
```

# Problem 2: Employee ID Card System

```python
Employee_Name = "Rahul"
Employee_ID = "EMP102"
Department = "IT"

class employee_id_card:
    def __init__(self, employee_name, employee_id, department):
        self.employee_name = employee_name
        self.employee_id = employee_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.employee_name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")

e1 = employee_id_card(Employee_Name, Employee_ID, Department)
e1.display_card()
```

# Problem 3: Mobile Contact Record

```python
Contact_Name = "Anita"
Phone_Number = 9876543210

class mobile_contact_record:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

m1 = mobile_contact_record(Contact_Name, Phone_Number)
m1.display_contact()
```
