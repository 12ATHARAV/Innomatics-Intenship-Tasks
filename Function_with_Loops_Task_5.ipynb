# Problem 1: Social Media – Post Engagement Analyzer

```python
likes_data = [200,123,345,120,190,420,235]

def post_engagement_analyzer(likes_data):
    total_likes = sum(likes_data)

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:",total_likes)
    print("Post Status:",status)


post_engagement_analyzer(likes_data)
```

# Problem 2: Healthcare – Medicine Stock Alert System

```python
stock = {"Paracetomol":5, "Crocine":12, "Montelucast":9, "Leocetrizine":15}

def medicine_stock_levels(stock):
    for medicine, quantity in stock.items():
        if quantity < 10:
            print(f"{medicine} is running low.\nQuantity: {quantity}\n")
        else:
            print(f"{medicine} is in stock.\nQuantity: {quantity}\n")

medicine_stock_levels(stock)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker


```python
data = {"May":23, "June":26, "July":29, "August":31, "Septeber":28}
required = 25

def seasonal_rainfall_tracker(data):
    average_rainfall = sum(data.values()) / len(data.values())
    if average_rainfall >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average_rainfall)
    print("Rainfall Status:", status)

seasonal_rainfall_tracker(data)
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
crop_data = {"Cotton":2100, "Tobacoo": 2700, "Soyabeans": 1700}

def premium_crop_filter(crop_data):
    lst = []
    for crop, price in crop_data.items():
        if price >= 2000:
            lst.append((crop,price))
    print("Premium Crops:",lst)

premium_crop_filter(crop_data)
```

# Problem 7: System Monitoring – Application Health Checker


```python
error_lst = ['error','error','na','na','na','na']

def application_health_checker(error_lst):
    error_count = error_lst.count('error')
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)


application_health_checker(error_lst)
```

# Problem 9: E-Learning – Student Attendance Eligibility System

```python
attendance_data = ['P','P','A','P','P','P']

def student_attendance(attendance_data):
    total_days = len(attendance_data)
    present_days = attendance_data.count('P')
    attendance_percentage = ((present_days / total_days) * 100).__round__(2)

    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)

student_attendance(attendance_data)
```

# Problem 10: Smart Electricity Bill Analyzer


```python
units_used = int(input("Enter the number of units used: "))

def smart_electricity_bill_analyzer(units_used):

    # bill calculator
    if units_used <= 100:
        bill = units_used * 3
    elif units_used <= 200:
        bill = (100 * 3) + ((units_used - 100) * 5)
    elif units_used > 200:
        bill = (100 * 3) + (100 * 5) + ((units_used - 200) * 7)

    # usage classification
    if bill < 500:
        usage = "Low"
    elif bill <= 1500:
        usage = "Moderate"
    else:
        usage = "High"

    print("Electricity Bill:", bill)
    print("Usage Classification:", usage)

smart_electricity_bill_analyzer(units_used)
```
