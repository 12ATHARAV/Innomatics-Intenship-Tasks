# Problem 1: Mobile Recharge Validation System


```python
def valid_recharge_system():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter the recharge amount: "))

        if amount < 50:
            print("Minimum recharge should be 50.")
        elif amount not in valid_plans:
            print("Invalid recharge amount.")
        else:
            print("Recharge successful.")
            break

valid_recharge_system()
```

# Problem 2: Inventory Reorder Alert System


```python
products = {"Oil":4,
            "Sugar": 23,
            "Coconut": 16,
            "Sesame": 14}

def inventory_system(products):
    for item, quantity in products.items():
        if quantity < 15:
            print(item,": Reorder Alert")
        else:
            print(item,": Stock OK")

inventory_system(products)
```

#  Problem 3: Student Result Processing System


```python
marks = [98,99,97,96,100]

def result_processing(marks):
    avg_marks = sum(marks) / len(marks)
    if avg_marks >= 50:
        print("Pass")
    else:
        print("Fail")

result_processing(marks)
```

# Problem 4: Cab Fare Estimator with Retry Option


```python
def cab_fare_estimator():
    while True:
        base_fare = 50
        cost_per_km = 12

        distance = float(input("Enter distance in km: "))
        total_fare_price = base_fare + (cost_per_km * distance)

        peak_hour = input("Is it peak hour? (yes / no): ")

        if peak_hour.lower() == "yes":
            total_fare_price += total_fare_price * 0.25
        elif peak_hour.lower() == "no":
            pass
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")
            continue

        print("Total Fare: ₹", total_fare_price)

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you!")
            break


cab_fare_estimator()
```

# Problem 5: Employee Attendance Eligibility Checker

```python
attendence = ['P','P','A','P','P','A','P','P','P']

def attendence_eligibility(lst):
    total_days = len(lst)
    present_days = lst.count('P')
    total_attendence = (present_days / total_days) * 100
    print("Your attendence: ", total_attendence.__round__(2),"%")

    if total_attendence >= 75:
        print("Eligible")
    else:
        print("Not Eligible")

attendence_eligibility(attendence)
```

# Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password  (Minimum 8 characters required)"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"


check_password_strength("123asJ34d@")
```
