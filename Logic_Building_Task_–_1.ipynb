## 1. User Login Check

### Problem Statement  
Given a username and password, check whether login is successful.

### Given Values
username = "admin"


password = "1234"


```python
username = "admin"
password = "1234"

input_username = input("Enter your username: ")
input_password = input("Enter your password: ")

if input_username == username and input_password == password:
    print("Login successfull")
else:
    print("Invalid credentials")
```

## 2. Pass / Fail Analyzer
### Problem Statement:
Analyze student results.


marks = [45, 78, 90, 33, 60]



### Requirements:
A student passes if marks â‰¥ 50


Count the total number of pass students


Count the total number of fail students


Print the final result clearly


```python
marks = [45, 78, 90, 33, 60]

def pass_fail_analyzer(lst):
    pass_students = 0
    fail_students = 0
    for mark in lst:
        if mark >= 50:
            pass_students += 1
        else:
            fail_students += 1
    return f"Passed Students: {pass_students}\nFailed Students: {fail_students}"

print(pass_fail_analyzer(marks))
```

## 3. Simple Data Cleaner
### Problem Statement:
Clean and standardize user names.


names = [" Alice ", "bob", " CHARLIE "]

### Requirements:
Remove extra spaces


Convert all names to lowercase


Print the cleaned list


```python
names = [" Alice ", "bob", " CHARLIE "]

def data_cleaner(lst):
    cleaned_list = []
    for name in lst:
        cleaned_name = name.strip().lower()
        cleaned_list.append(cleaned_name)
    return cleaned_list

data_cleaner(names)
```

## 4. Message Length Analyzer
### Problem Statement:
Analyze message sizes.


messages = ["Hi", "Welcome to the platform", "OK"]

### Requirements:
Print the length of each message


Flag messages longer than 10 characters


```python
messages = ["Hi", "Welcome to the platform", "OK"]

def length_analyzer(lst):
    for message in lst:
        print(f"Message: {message}, Length: {len(message)}")
        if len(message) > 10:
            print("This message is longer than 10 characters")

        print()    #blank line for readability

length_analyzer(messages)
```

## 5. Error Message Detector
### Problem Statement:
Detect error messages from system logs.


logs = ["INFO", "ERROR", "WARNING", "ERROR"]

### Requirements:
Count the number of "ERROR" entries


Print the total error count


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

def error_messages(lst):
    count = 0
    for log in lst:
        if log == "ERROR":
            count += 1
    return f"Total Errors: {count}"

error_messages(logs)
```
