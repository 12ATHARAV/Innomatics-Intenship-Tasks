```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile"  
)
```

## Task 1: PromptTemplate

```python
from langchain_core.prompts import PromptTemplate

template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

prompt = template_1.format(topic="Artificial Intelligence")

response = llm.invoke(prompt)
print(response.content)
```

## Task 2: Multi-Input Prompt

```python
multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

prompt = multi_template.format(
    topic="Python",
    audience="kids",
    tone="fun"
)

print(llm.invoke(prompt).content)
```

## Task 3: Prompt Variations Engine

```python
teaching = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(llm.invoke(teaching.format(topic=topic)).content)
print(llm.invoke(interview.format(topic=topic)).content)
print(llm.invoke(story.format(topic=topic)).content)
```

## Task 4: ChatPromptTemplate

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

messages = chat_template.format_messages(
    role="teacher",
    topic="Neural Networks"
)

response = llm.invoke(messages)
print(response.content)
```

## Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone
```

## Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling way"
    }

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=styles[style]
    )

    prompt = template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    return llm.invoke(prompt).content
```

```python
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
reuse_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

topics = ["AI", "ML", "Python", "Data Science", "Deep Learning"]

for t in topics:
    prompt = reuse_template.format(topic=t)
    print(llm.invoke(prompt).content)
    print("-----------------------")
```
